```python
import os 
import matplotlib.pyplot as plt 
import pandas as pd 
import seaborn as sns 
import csv
```

```python
df = pd.read_csv('combined.csv', sep=',', usecols=['Identity (%)', 'SampleName', 'GeneId', 'competency', 'SourceFile'])

x_axis_order =['Standard', 'Low-Quality', 'Very-Poor-Quality', 'Low-Read-Length', 'Low-Depth']

plt.figure(figsize=(12,6))
sns.boxplot(x='SourceFile', y='Identity (%)', data=df, color='skyblue', showfliers=False, order=x_axis_order)
#sns.stripplot(x='SourceFile', y='Identity (%)', data=df, hue='competency', size=5, jitter=0.35, alpha=0.1, dodge=True, linewidth=1)

plt.xlabel('Synthetic Read Quality')
plt.ylabel('Alingment Rate (%)')
plt.legend(title="Isolate", bbox_to_anchor=(1, 1), ncol=5)
```
